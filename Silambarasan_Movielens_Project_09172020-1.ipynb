# 8. Visualize how popularity of genres has changed over the years

```python
# Creating new dataframe by copying  Movie_Data Dataframe- 
# Apply groupby and make 'release year' as index and 'all genres' as columns 
# Apply sum() to get the number of movies released per genre in a year and dropping other columns

Genre_temp=Movie_Data.copy()
Genre_Popularity=Genre_temp.groupby('release year')[Genre_temp.columns].sum()
Genre_Popularity.drop(columns=['release year','movie id'],axis=1,inplace=True)
Genre_Popularity.tail()
```

```python
# Creating new column to store total number of movies released in all genre per year

Genre_Popularity['Total']=Genre_Popularity.sum(axis=1)
```

```python
Genre_Popularity.tail(5)

#New column Total is added in datadrame
```

```python
#Calculating the percentage of number of release per genre in a year

Genre_Popular_Percentage_year=Genre_Popularity.div(Genre_Popularity['Total'], axis= 0) * 100

#Dropping columns - 
#   1. Total - created for computation purpose only
#   2. unknown genre - Currently 'unknown genre'=0

Genre_Popular_Percentage_year.drop(columns=['Total','unknown'],axis=1,inplace=True)

print("Genres Vs Most popular Year \n")
print(Genre_Popular_Percentage_year.idxmax())
```

```python
# Viewing last 5 rows
Genre_Popular_Percentage_year.tail()
```

```python
# Univariate Analysis - Plotting the popularity of genres over year

plt.figure(figsize=(12,8))
sns.set_style("darkgrid")
Action = plt.plot(Genre_Popular_Percentage_year['Action'],color='b')
Adventure = plt.plot(Genre_Popular_Percentage_year['Adventure'],color='g')
Animation = plt.plot(Genre_Popular_Percentage_year['Animation'],color='r')
Childrens = plt.plot(Genre_Popular_Percentage_year['Childrens'],color="c")
Comedy = plt.plot(Genre_Popular_Percentage_year['Comedy'],color="m")
Crime = plt.plot(Genre_Popular_Percentage_year['Crime'], color ="y")
Documentary = plt.plot(Genre_Popular_Percentage_year['Documentary'],color='lime')
Drama = plt.plot(Genre_Popular_Percentage_year['Drama'],color='mediumpurple')
Fantasy = plt.plot(Genre_Popular_Percentage_year['Fantasy'],color='yellowgreen')
FilmNoir = plt.plot(Genre_Popular_Percentage_year['Film-Noir'],color="olive")
Horror = plt.plot(Genre_Popular_Percentage_year['Horror'],color="aqua")
Musical = plt.plot(Genre_Popular_Percentage_year['Musical'],color="teal")
Mystery = plt.plot(Genre_Popular_Percentage_year['Mystery'],color='turquoise')
Romance = plt.plot(Genre_Popular_Percentage_year['Romance'],color='brown')
SciFi = plt.plot(Genre_Popular_Percentage_year['Sci-Fi'],color='black')
Thriller = plt.plot(Genre_Popular_Percentage_year['Thriller'],color='indigo')
War = plt.plot(Genre_Popular_Percentage_year['War'],color='darkorange')
Western = plt.plot(Genre_Popular_Percentage_year['Western'],color='gold')
plt.title('Popularity of Genres over years')
plt.xlabel('Year')
plt.ylabel('%of Genre')
plt.legend(['Action','Adventure','Animation','Children','Comedy','Crime','Documentary','Drama','Fantasy','Film-Noir','Horror','Musical','Mystery','Romance','Sci-Fi','Thriller','War','Western'], loc=1)
plt.show()
```

####   Drama Genre is  currently popular than all other genres

# 9. Find the top 25 movies according to average ratings such that each movie has number of ratings more than 100

```python
# Merging two dataframe with column Movie id and condition - inner join

Merge_Dataframe = pd.merge(User_Rating,Movie_Data, how='inner',on='movie id')
Merge_Dataframe.shape
```

```python
# Checking for null value after merge

Merge_Dataframe.isna().apply(pd.value_counts)
```

```python
# Viewing first 5 rows

Merge_Dataframe.head()
```

```python
# Creating new dataframe with needed columns from Merge_Dataframe

M=Merge_Dataframe[['movie title','rating']]
```

```python
# Applying groupby - column: movie title
# Apply aggregate function on column rating to get overall number of ratings and overall average rating for a movie

Top25Movies=M.groupby('movie title').rating.agg(['sum','mean'])

print("\tTop 25 Movies   \t\t    # of Rating>100 Avg_Rating    \t\n")
print(Top25Movies[Top25Movies['sum']>100].sort_values(by='mean',ascending=False).head(25))
```

```python
# Creating new dataframe to find total number of ratings for male and female

Rating =Comp_df[['gender','rating']]
Rating.head()
```

```python
# Applying groupby on column gender
# Get the total number of rating for both male and female using sum() function
# Rename the column name and display df

Rating_temp= Rating.groupby(['gender'])['rating'].sum()
Rating_df= pd.DataFrame(Rating_temp)
Rating_df.columns=['Total_Rating']
Rating_df
```

```python
# Creating copy of comp_df dataframe
# Create new dataframe with needed columns

Comp_df_temp=Comp_df.copy()
temp1=Comp_df[['movie title','rating','gender','Drama','Sci-Fi','Romance']]
```

```python
# Viewing the first 5 rows

temp1.head()
```

```python
# Applying Groupby on columns - Drama and Gender 
# Get the total number of rating  for Drama genre per gender  Using sum() function
# Reset the index and dropping rows with Drama=0

Drama_Rating_temp= temp1.groupby(['Drama','gender'])['rating'].sum()
Drama_Rating=Drama_Rating_temp.reset_index().drop([0,1],axis=0)
Drama_Rating_df= pd.DataFrame(Drama_Rating)
Drama_Rating_df
```

```python
# Dropping the column-Drama
# Rename the column names

Drama_Rating_df.drop('Drama',axis=1,inplace=True)
Drama_Rating_df.columns=['gender','Drama_Rating']
Drama_Rating_df
```

```python
# Set the index as gender
# Displaying the dataframe

a=Drama_Rating_df.set_index('gender')
a
```

```python
# Applying Groupby on columns - Sci-Fi and Gender 
# Get the total number of rating  for Sci-Fi per Gender Using sum() function
# Reset the index and dropping rows with Sci-Fi=0

SciFi_Rating_temp= temp1.groupby(['Sci-Fi','gender'])['rating'].sum()
SciFi_Rating_temp
SciFi_Rating= SciFi_Rating_temp.reset_index().drop([0,1],axis=0)
SciFi_Rating_df=pd.DataFrame(SciFi_Rating)
SciFi_Rating_df
```

```python
# Dropping the  column - Sci-Fi
# Renaming the columns

SciFi_Rating_df.drop('Sci-Fi',axis=1,inplace=True)
SciFi_Rating_df.columns=['gender','SciFi_Rating']
SciFi_Rating_df
```

```python
# Set the index as gender and display df

b=SciFi_Rating_df.set_index('gender')
b
```

```python
# Applying Groupby on columns - Romance and Gender 
# Get the total number of rating  for Romance per Gender Using sum() function

Romance_Rating_temp= temp1.groupby(['Romance','gender'])['rating'].sum()
Romance_Rating_temp
```

```python
# Reset the index and dropping rows with Romance=0

Romance_Rating= Romance_Rating_temp.reset_index().drop([0,1],axis=0)
Romance_Rating_df=pd.DataFrame(Romance_Rating)
Romance_Rating_df
```

```python
# Dropping the column -Romance
# Renaming the column

Romance_Rating_df.drop('Romance',axis=1,inplace=True)
Romance_Rating_df.columns=['gender','Romance_Rating']
Romance_Rating_df
```

```python
# Set the index as gender and  display df

c=Romance_Rating_df.set_index('gender')
c
```
